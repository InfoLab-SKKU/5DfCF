```python
import pandas as pd

file_path = 'data/csv_data/ADNIMERGE_12Dec2024.csv'
data = pd.read_csv(file_path)

sorted_data = data.sort_values(by='PTID')
sorted_data
```

```python
filtered_data = sorted_data[['RID', 'PTID', 'EXAMDATE', 'AGE', 'DX']]
filtered_data
```

```python
filtered_data = filtered_data[filtered_data['DX'].isin(['MCI', 'Dementia'])]
filtered_data
```

```python
# Define a function to assign labels and filter data
def assign_label_and_filter(group):
    unique_dx = group['DX'].unique()
    if len(unique_dx) == 1 and unique_dx[0] == 'Dementia':
        return None  # Mark this group for removal
    elif len(unique_dx) == 1 and unique_dx[0] == 'MCI':
        group['label'] = 0
    elif 'MCI' in unique_dx and 'Dementia' in unique_dx:
        group['label'] = 1
    return group

# Apply the function to each PTID group
processed_data = (
    filtered_data.groupby('PTID')
    .apply(assign_label_and_filter)
    .reset_index(drop=True)
)

# Remove groups marked as None
processed_data = processed_data.dropna()
processed_data
```

```python
# Convert EXAMDATE to datetime for accurate sorting
processed_data['EXAMDATE'] = pd.to_datetime(processed_data['EXAMDATE'])

# Sort within each PTID group by EXAMDATE
processed_data = processed_data.sort_values(by=['PTID', 'EXAMDATE'])
processed_data
```

```python
# Save the processed data to a new CSV file
output_path = 'data/csv_data/new_adni_processed/adnimerged_with_labels_sorted_by_date_not_changed_age.csv'
processed_data.to_csv(output_path, index=False)
```

```python
# Add a new column 'current_age' based on baseline AGE and EXAMDATE
def calculate_current_age(group):
    baseline_date = group.iloc[0]['EXAMDATE']
    baseline_age = group.iloc[0]['AGE']
    group['current_age'] = group['EXAMDATE'].apply(
        lambda x: round(baseline_age + (x - baseline_date).days / 365.25, 1)
    )
    return group

processed_age_data = processed_data.groupby('PTID').apply(calculate_current_age)
processed_age_data
```

```python
# Save the processed data to a new CSV file

output_path = 'data/csv_data/new_adni_processed/adnimerged_with_labels_sorted_by_date_changed_age.csv'
processed_age_data.to_csv(output_path, index=False)
processed_age_data
```

```python
import pandas as pd

file_path_1 = 'data/csv_data/new_adni_processed/adnimerged_with_labels_sorted_by_date_changed_age.csv'
file_path_2 = 'data/csv_data/adni_csv_fmri_processed/adni_csv_fmri_date_changed.csv'

data_1 = pd.read_csv(file_path_1)
data_2 = pd.read_csv(file_path_2)

print("Data 1 preview:")
print(data_1.head())

print("\nData 2 preview:")
print(data_2.head())

```

```python
data_2.rename(columns={'Subject': 'PTID'}, inplace=True)
data_2
```

```python

data_1['EXAMDATE'] = pd.to_datetime(data_1['EXAMDATE'])
data_2['Acq Date'] = pd.to_datetime(data_2['Acq Date'])

data_1['EXAM_YEAR_MONTH'] = data_1['EXAMDATE'].dt.to_period('M')
data_2['ACQ_YEAR_MONTH'] = data_2['Acq Date'].dt.to_period('M')

merged_data = pd.merge(
    data_1,
    data_2,
    left_on=['PTID', 'EXAM_YEAR_MONTH'],
    right_on=['PTID', 'ACQ_YEAR_MONTH'],
    how='inner'
)
merged_data
```

```python
unmerged_data = data_2[~data_2.index.isin(merged_data.index)]
unmerged_data
```

```python
label_0_count = merged_data[merged_data['label'] == 0]['PTID'].nunique()

label_1_count = merged_data[merged_data['label'] == 1]['PTID'].nunique()
print(f"Number of unique PTIDs with label 0: {label_0_count}")
print(f"Number of unique PTIDs with label 1: {label_1_count}")
```

```python
merged_data['EXAM_YEAR_MONTH'] = merged_data['EXAM_YEAR_MONTH'].astype(str)
merged_data['EXAM_YEAR_MONTH'] = pd.to_datetime(merged_data['EXAM_YEAR_MONTH'])

def filter_within_four_years_and_one_year_intervals(group):
    group = group.sort_values(by='EXAM_YEAR_MONTH', ascending=False)
    filtered_rows = []
    last_date = None
    latest_date = group['EXAM_YEAR_MONTH'].iloc[0]

    for _, row in group.iterrows():
        current_date = row['EXAM_YEAR_MONTH']
        if current_date < latest_date - pd.DateOffset(years=4):
            break
        if last_date is None or (last_date - current_date).days >= 365:
            filtered_rows.append(row)
            last_date = current_date

    return pd.DataFrame(filtered_rows)

filtered_data = merged_data.groupby('PTID').apply(filter_within_four_years_and_one_year_intervals).reset_index(drop=True)

filtered_data
```

```python

def fill_missing_years_with_copy(group):
    group = group.sort_values(by='EXAM_YEAR_MONTH', ascending=False).reset_index(drop=True)

    group['YEAR'] = group['EXAM_YEAR_MONTH'].astype(str).str[:4].astype(int)
    years = group['YEAR']
    max_year = years.max()
    years = years.tolist()
    expected_years = list(range(max_year, max_year - 5, -1))

    missing_years = [year for year in expected_years if year not in years]

    print(f"Processing PTID: {group['PTID'].iloc[0]}")
    print(f"Existing years: {years}")
    print(f"Expected years: {expected_years}")
    print(f"Missing years: {missing_years}")

    result = group.copy()

    for year in missing_years:
        print(f"Filling missing year {year} for PTID {group['PTID'].iloc[0]}")

        previous_rows = result[result['YEAR'] > year]
        if not previous_rows.empty:
            latest_row = previous_rows.iloc[-1]
            print(f"Selected row to copy for year {year}:\n{latest_row}")

            new_row = latest_row.copy()
            new_row['EXAM_YEAR_MONTH'] = latest_row['EXAM_YEAR_MONTH'].replace(year=year)
            new_row['YEAR'] = year

            result = pd.concat([result, new_row.to_frame().T], ignore_index=True)

    result = result.sort_values(by='EXAM_YEAR_MONTH', ascending=False).reset_index(drop=True)
    return result.drop(columns=['YEAR'])

filtered_data_with_copy = (
    filtered_data_without_one_entry.groupby('PTID')
    .apply(fill_missing_years_with_copy)
    .reset_index(drop=True)
)

filtered_data_with_copy
```
